## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
cur.execute("SELECT DISTINCT Launch_Site FROM users")
for i in cur.fetchall():
    print(i[0])
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
cur.execute("SELECT * FROM users WHERE Launch_Site LIKE 'CCA%' LIMIT 5;")
for i in cur.fetchall():
    print(i[0])
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
cur.execute("SELECT SUM(PAYLOAD_MASS__KG_) total FROM users;")
for i in cur.fetchall():
    print(i[0])

```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
cur.execute("SELECT AVG(PAYLOAD_MASS__KG_) FROM users WHERE Booster_Version LIKE 'F9 v1.1%'")
for i in cur.fetchall():
    print(i[0])
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
cur.execute("SELECT Date From users ORDER BY Date Limit 1;")
for i in cur.fetchall():
    print(i[0])
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
cur.execute("SELECT Booster_Version,PAYLOAD_MASS__KG_ From users WHERE Mission_Outcome LIKE 'Success' AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000;")
for i in cur.fetchall():
    print(str(i[0]) + " " + str(i[1]))
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
cur.execute("SELECT Mission_Outcome, COUNT(DISTINCT Mission_Outcome) FROM users;")
for i in cur.fetchall():
    print(i[0] + " " + str(i[1]))
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
cur.execute("SELECT Booster_Version FROM users WHERE PAYLOAD_MASS__KG_ == (SELECT max(PAYLOAD_MASS__KG_) FROM users);")
for i in cur.fetchall():
    print(i[0])
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
cur.execute("SELECT Booster_Version, Date FROM users WHERE 'Landing _Outcome' != 'Success%' AND Date LIKE '%-2015'")
for i in cur.fetchall():
    print(i[0]+ " " + str(i[1]))




```
